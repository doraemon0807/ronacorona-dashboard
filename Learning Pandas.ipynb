```python
import pandas as pd

daily_df = pd.read_csv("data/daily_report.csv")

deaths_df = pd.read_csv("data/time_deaths.csv")
recovered_df = pd.read_csv("data/time_recovered.csv")
```

```python
totals_df = daily_df[["Confirmed", "Deaths", "Recovered"]].sum()

totals_df = totals_df.reset_index(name="count")
totals_df = totals_df.rename(columns={"index":"condition"})

totals_df
```

```python
countries_df = daily_df[["Country_Region", "Confirmed", "Deaths", "Recovered"]]

countries_df = countries_df.groupby("Country_Region").sum().reset_index()

countries_df
```

```python
conditions = ["confirmed", "deaths", "recovered"]


def make_df(condition, country=None):
    df = pd.read_csv(f"data/time_{condition}.csv") 
    if country is not None:
        df = df.loc[df["Country/Region"] == country] 
    df = df.drop(columns=["Province/State", "Country/Region", "Lat", "Long"]).sum().reset_index(name=condition) 
    df = df.rename(columns={"index": "date"}) 
    return df

def make_global_df():
    final_df = None
    for condition in conditions:
        condition_df = make_df(condition)
        if final_df is None:
            final_df = condition_df
        else:
            final_df = final_df.merge(condition_df)
            
    return final_df
        
global_df = make_global_df()
global_df
```

```python
def make_country_df(country): 
    final_df = None 
    for condition in conditions: 
        condition_df = make_df(condition, country) 
        if final_df is None: 
            final_df = condition_df 
        else: 
            final_df = final_df.merge(condition_df)            
    return final_df

df = make_country_df("Iran")
df
```
